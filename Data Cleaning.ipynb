# Cleaning p-value Dataset

```python
data1.head(50)
```

```python
extracted_list_columns = list(extracted.keys())
for i in range(len(extracted_list_columns)):
    extracted_list_columns[i] = extracted_list_columns[i].upper()

names = ['cancer', 'diabetes', 'smoke_often', 'smoke_rarely', 'never_smoked', 'left_hand', 'right_hand', 'cat', 'dog']

cleaned = data1[data1['food'].isin(extracted_list_columns)]
removed_char = cleaned[cleaned['characteristic'].isin(names)]
#create significant column
def significance_label(p_value):
    if p_value < 0.05:
        return "significant"
    else:
        return "not significant"

# Apply the function to the 'p_values' column to create the 'significant' column
removed_char["significant"] = removed_char["p_values"].apply(significance_label)
# remove_char["significant"] = [removed_char["p_values"] < 0.1
sig = removed_char[removed_char["p_values"] < 0.1]
sig["significant"] = sig["p_values"].apply(significance_label)
sig['food_name'] = sig['food'].map(extracted)



print(data1.shape)
print(cleaned.shape)
print(removed_char.shape)
print(sig.shape)

print(sig["characteristic"].value_counts(""))
        
sig.head()
```
